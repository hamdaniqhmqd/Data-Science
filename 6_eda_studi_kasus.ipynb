### 27. Hapus baris-baris yang mengandung missing values dalam kolom 'SalePrice'.

```python
df = house_price_train.dropna(subset=['SalePrice'])
```

### 28. Lakukan imputasi pada sebuah kolom lainnya (bebas) yang mengandung missing values dengan strategi ‘median’.

```python
df['LotFrontage'] = df['LotFrontage'].fillna(df['LotFrontage'].median())
```

### 29. Visualkan kembali untuk memeriksa kolom-kolom yang masih memiliki missing values.


```python
# Visualisasi missing values setelah imputasi
msno.matrix(df)
plt.show()

# Menampilkan kolom yang masih memiliki missing values
df.isnull().sum()
```

Visualisasi missing values setelah imputasi menunjukkan kolom-kolom yang masih memiliki nilai hilang, membantu memastikan bahwa proses imputasi telah berhasil mengurangi atau menghilangkan missing values. Namun pada kolom data yang tampil pada hasil fungsi `.isnull().sum()` menunjukan bahwa kolom tersebut tidak memiliki `missing value`.

### 30. Tampilkan visualisasi heatmap untuk kolom-kolom yang saling berkaitan. Tampilkan chart-chart lainnya untuk memvisualisasi sebaran dan keterkaitan antar variable pada dataset (Univariate/Bivariate/Multivariate).

```python
# 1. Heatmap untuk melihat korelasi antar kolom numerik
plt.figure(figsize=(12, 8))
# Memilih hanya kolom numerik untuk heatmap
numeric_df = df.select_dtypes(include=['number'])
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title('Heatmap Korelasi Kolom Numerik')
plt.show()

# 2. Contoh visualisasi univariate (distribusi satu variabel)
plt.figure(figsize=(8, 6))
sns.histplot(df['SalePrice'], kde=True, color='blue')
plt.title('Distribusi SalePrice')
plt.xlabel('SalePrice')
plt.ylabel('Frekuensi')
plt.show()

# 3. Contoh visualisasi bivariate (hubungan dua variabel)
plt.figure(figsize=(8, 6))
sns.scatterplot(x='GrLivArea', y='SalePrice', data=df, color='green')
plt.title('Hubungan GrLivArea dan SalePrice')
plt.xlabel('GrLivArea')
plt.ylabel('SalePrice')
plt.show()

# 4. Contoh visualisasi multivariate (hubungan lebih dari dua variabel)
plt.figure(figsize=(10, 8))
sns.pairplot(df[['SalePrice', 'GrLivArea', 'TotalBsmtSF', 'OverallQual']])
plt.suptitle('Pairplot untuk Multivariate Analysis', y=1.02)
plt.show()
```

Heatmap dan chart lainnya (seperti histogram, scatter plot, dan pair plot) digunakan untuk memvisualisasikan korelasi, distribusi, dan hubungan antar variabel dalam dataset, memberikan insight tentang pola dan keterkaitan data.
