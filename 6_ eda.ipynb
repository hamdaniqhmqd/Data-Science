### 5. Buat sebuah fungsi untuk menghitung jumlah dan persentase missing values dalam setiap kolom suatu dataframe, serta menyajikan informasi tersebut dalam bentuk tabel.



```python
def missing_values_table(df):
    # Total missing values
    mis_val = df.isnull().sum()

    # Percentage of missing values
    mis_val_percent = 100 * df.isnull().sum() / len(df)

    # Make a table with the results
    mis_val_table = pd.concat([mis_val, mis_val_percent], axis=1)

    # Rename the columns
    mis_val_table_ren_columns = mis_val_table.rename(
        columns={0: 'Missing Values', 1: '% of Total Values'}
    )

    # Sort the table by percentage of missing values in descending order
    mis_val_table_ren_columns = mis_val_table_ren_columns[
        mis_val_table_ren_columns.iloc[:, 1] != 0
    ].sort_values('% of Total Values', ascending=False).round(1)

    # Print summary information
    print(f"Your selected dataframe has {df.shape[1]} columns.\n"
          f"There are {mis_val_table_ren_columns.shape[0]} columns that have missing values.")

    # Return the dataframe with missing information
    return mis_val_table_ren_columns
```

Fungsi ini digunakan untuk menghitung jumlah dan persentase nilai yang hilang `(missing values)` dalam setiap kolom dari sebuah dataframe df, lalu menyajikannya dalam bentuk tabel yang telah diurutkan berdasarkan persentase nilai yang hilang dari yang tertinggi ke terendah.

### 15. Pairwise Deletion (penghapusan berpasangan). Digunakan ketika nilai hilang sepenuhnya secara acak `(MCAR)`. Selama penghapusan berpasangan, hanya nilai yang hilang yang dihapus. Semua operasi seperti mean, sum, dll. mengabaikan nilai yang hilang.


```python
titanic_1 = titanic.copy()
titanic_1['Age'].mean()
```

Nilai `np.float64(29.69911764705882)` adalah hasil perhitungan rata-rata `(mean)` dari kolom Age dalam dataset `titanic_1`, yang menunjukkan bahwa rata-rata usia penumpang dalam dataset tersebut adalah sekitar `29.7` tahun.

### 16. `Listwise Deletion`. Baris lengkap (yang berisi nilai yang hilang) akan dihapus. Disebut juga Complete Case Deletion. Seperti `Pairwise Deletion`, `Listwise Deletion` juga hanya digunakan untuk nilai `MCAR`.

```python
titanic_1.dropna(subset=['Age'], how='any', inplace=True)
```

berfungsi untuk menghapus semua baris dalam dataframe `titanic_1` yang memiliki missing value (NaN) pada kolom Age.

## C. Teknik Imputasi


### 18. Teknik imputasi mengacu pada penggantian data yang hilang dengan suatu nilai. Teknik imputasi secara garis besar dapat diklasifikasikan sebagai berikut:


- Imputasi menggunakan statistik (rata-rata, median, atau most frequent) dari
setiap kolom tempat nilai yang hilang berada.
- Imputasi dengan nilai konstan.

### 19. Mengamputasi dengan nilai konstan menggunakan library `scikit-learn`

install dan gunakan library tersebut

```python
from sklearn.impute import SimpleImputer
titanic_most_frequent = titanic.copy()

mean_inputer = SimpleImputer(strategy='most_frequent')
titanic_most_frequent.iloc[:,:] = mean_inputer.fit_transform(titanic_most_frequent)
```

### 20. Tampilkan informasi untuk memastikan bahwa semua variabel yang memiliki missing value sudah tidak ada lagi (0).

```python
msno.matrix(titanic_most_frequent)
```

Hasil tersebut menunjukan bahwa tidak ada missing value pada setiap kolom data.

### 21. Ganti strategi imputasi dengan metode `'constant'` dengan fill_value=0.


```python
# Mengubah strategi imputasi dengan fill_value=0
imputer_zero = SimpleImputer(strategy='constant', fill_value=0)

# Melakukan imputasi ulang dengan nilai 0
titanic_most_frequent[:] = imputer_zero.fit_transform(titanic)

titanic_most_frequent.isnull().sum()
```

Seluruh kolom dalam dataset sudah tidak memiliki nilai yang hilang (missing values) setelah dilakukan imputasi.
