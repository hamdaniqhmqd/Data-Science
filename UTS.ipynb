```python
# Pastikan ada kolom 'Category' yang membedakan antara jenis produk
if 'Category' in supermarket.columns:
    # Hitung jumlah produk per kategori dalam setiap State
    Furniture = supermarket[supermarket['Category'] == 'Furniture']['State'].value_counts().sort_index()
    Office_Supplies = supermarket[supermarket['Category'] == 'Office Supplies']['State'].value_counts().sort_index()
    Technology = supermarket[supermarket['Category'] == 'Technology']['State'].value_counts().sort_index()

    # Urutkan berdasarkan alfabet (A-Z)
    sorted_states = sorted(set(supermarket['State']))  # Dapatkan daftar State dalam urutan abjad

    # Pastikan semua State ada dalam setiap kategori (isi 0 jika tidak ada)
    Furniture = Furniture.reindex(sorted_states, fill_value=0)
    Office_Supplies = Office_Supplies.reindex(sorted_states, fill_value=0)
    Technology = Technology.reindex(sorted_states, fill_value=0)

    # Plot line chart
    plt.figure(figsize=(14, 6))
    plt.plot(Furniture.index, Furniture.values, marker='o', linestyle='-', label='Furniture', color='blue')
    plt.plot(Office_Supplies.index, Office_Supplies.values, marker='o', linestyle='-', label='Office Supplies', color='red')
    plt.plot(Technology.index, Technology.values, marker='o', linestyle='-', label='Technology', color='yellow')

    plt.xlabel('State')
    plt.ylabel('Jumlah Produk')
    plt.title('Perbandingan Jumlah Produk per Kategori di Setiap State')
    plt.xticks(rotation=90)
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.7)

    plt.show()
```
