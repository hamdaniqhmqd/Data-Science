### 38. Menampilkan scatter plot hubungan antar kolom data fare (harga tiket) dan age (usia penumpang).

menampilkan hubungan antara kolom data fare dan age

```python
# menampilkan hubungan antara data fare & age
x_data = titanic["Fare"]
y_data = titanic["Age"]

plt.scatter(x_data, y_data)

plt.xlabel("Fare")
plt.ylabel("Age")
plt.title("Data sebaran Harga tiket sesuai usia")

plt.show()
```

### 39. Tampilkan grafik antara kelas dan jumlah penumpang yang selamat menggunakan bar chart.

Menampilkan kolom kelas dan jumlah penumpang yang selamat menggunakan grafik bar chart

```python
survived_per_class = titanic[titanic["Survived"] == 1].groupby("Pclass")["Survived"].count()

plt.figure(figsize=(8,5))
sns.barplot(x=survived_per_class.index, y=survived_per_class.values, palette="viridis")

plt.xlabel("Kelas Penumpang")
plt.ylabel("Jumlah Penumpang Selamat")
plt.title("Jumlah Penumpang Selamat berdasarkan Kelas")

plt.show()
```

### 40. Tampilkan grafik antara jenis kelamin dan jumlah penumpang yang selamat menggunakan pie chart

menampilkan kolom jenis kelamin dan jumlah penumpang, dengan bantuan pie chart

```python
# Menampilkan pie chart jumlah penumpang selamat berdasarkan jenis kelamin
survived_by_gender = titanic[titanic["Survived"] == 1].groupby("Sex")["Survived"].count()

plt.figure(figsize=(6,6))
plt.pie(survived_by_gender, labels=survived_by_gender.index, autopct="%1.1f%%", startangle=90, colors=['lightblue', 'pink'])
plt.title("Persentase Penumpang Selamat berdasarkan Jenis Kelamin")
plt.show()
```

### 41. Tampilkan grafik rata-rata usia penumpang dari Titanic per kelas menggunakan bar chart.

menampilkan data kolom rata-rata usia penumpang sesuai dengan kelasnya, dengan grafik bar chart

```python
# Menampilkan bar chart rata-rata usia penumpang berdasarkan kelas
plt.figure(figsize=(8,5))
sns.barplot(data=titanic, x="Pclass", y="Age", ci=None, palette="magma")

plt.xlabel("Kelas Penumpang")
plt.ylabel("Rata-rata Usia")
plt.title("Rata-rata Usia Penumpang berdasarkan Kelas")
plt.show()
```

### 42. Tampilkan grafik distribusi jumlah penumpang berdasarkan kelas

membandingkan data penumpang pada setiap kelasnya

```python
# Menampilkan distribusi jumlah penumpang berdasarkan kelas
plt.figure(figsize=(8,5))
sns.countplot(data=titanic, x="Pclass", palette="coolwarm")
plt.xlabel("Kelas Penumpang")
plt.ylabel("Jumlah Penumpang")
plt.title("Distribusi Jumlah Penumpang berdasarkan Kelas")
plt.show()
```

### 43. Buatlah kode untuk menunjukkan jumlah sibling/spouse pada penumpang Titanic

Histogram ini bertujuan untuk menampilkan distribusi jumlah sibling/spouse (saudara kandung/pasangan) yang dimiliki oleh setiap penumpang di Titanic. 

```python
# Histogram jumlah sibling/spouse pada penumpang Titanic
plt.figure(figsize=(8,5))
plt.hist(titanic["SibSp"], bins=8, color='skyblue', edgecolor='black')

plt.xlabel("Jumlah Sibling/Spouse")
plt.ylabel("Jumlah Penumpang")
plt.title("Histogram Sibling/Spouse")

plt.show()
```
