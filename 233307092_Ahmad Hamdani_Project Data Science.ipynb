### **6.6. Menghitung dan memvisualisasikan *Silhouette Score* untuk berbagai jumlah cluster guna validasi**

Tujuan dari kode ini adalah untuk mengevaluasi kualitas *clustering* yang dihasilkan oleh algoritma K-Means untuk berbagai jumlah cluster (dari 2 hingga 10) menggunakan metrik *Silhouette Score*. *Silhouette Score* mengukur seberapa mirip suatu objek dengan cluster-nya sendiri dibandingkan dengan cluster lain. Skor yang tinggi menunjukkan bahwa objek tersebut cocok dengan cluster-nya sendiri dan kurang mirip dengan cluster tetangga. Hasil *Silhouette Score* untuk setiap jumlah cluster kemudian divisualisasikan dalam bentuk *bar chart* untuk memudahkan perbandingan.

Hasil program mencetak *Silhouette Score* untuk setiap jumlah cluster yang diuji, dan menampilkan sebuah *bar chart* di mana tinggi setiap batang merepresentasikan *Silhouette Score* untuk jumlah cluster yang sesuai. Berdasarkan skor yang dicetak dan visualisasi, jumlah cluster 4 memberikan *Silhouette Score* tertinggi (sekitar 0.6179), yang menunjukkan bahwa *clustering* dengan 4 cluster memiliki pemisahan yang baik antar cluster. Ini mendukung hasil yang diperoleh dari metode Elbow sebelumnya.

```python
silhouette_scores = {}
range_n_clusters = range(2, 11)  # Coba jumlah cluster dari 2 hingga 10

for n_clusters in range_n_clusters:
    kmeans = KMeans(n_clusters=n_clusters, random_state=42, n_init=10)
    cluster_labels = kmeans.fit_predict(X)
    silhouette_avg = silhouette_score(X, cluster_labels)
    silhouette_scores[n_clusters] = silhouette_avg
    print(f"Silhouette Score untuk {n_clusters} Cluster: {silhouette_avg:.4f}")

# Visualisasikan Silhouette Scores
plt.figure(figsize=(10, 6))
plt.bar(silhouette_scores.keys(), silhouette_scores.values())
plt.xlabel("Jumlah Cluster")
plt.ylabel("Silhouette Score")
plt.title("Silhouette Score untuk Berbagai Jumlah Cluster")
plt.xticks(np.arange(min(silhouette_scores.keys()), max(silhouette_scores.keys()) + 1, 1.0))
plt.grid(True)
plt.show()
```

### **Kesimpulan**

1.  **Perokok Pasif - Kesehatan Parah:** Kelompok dengan usia cenderung lebih tua dan indikator kesehatan yang mengkhawatirkan meskipun tidak merokok.

3.  **Perokok Pasif - Kesehatan Tidak Parah:** Kelompok yang tidak merokok dengan indikator kesehatan yang relatif lebih baik dibandingkan kelompok perokok pasif lainnya.
4.  **Perokok Aktif - Kesehatan Parah:** Perokok aktif dengan usia relatif muda namun menunjukkan indikator kesehatan yang mengkhawatirkan.
