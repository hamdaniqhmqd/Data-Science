### 19. Membuat bar chart dengan multi values (stacked bar).

```python
years = ['2016', '2017', '2018', '2019']
rice = [57, 50, 47, 30]
wheat = [43, 50, 53, 70]

plt.figure(figsize=(10, 6))

plt.bar(years, rice, label='Rice')
plt.bar(years, wheat, bottom=rice, label='Wheat')

plt.xlabel('Years', fontsize=14)
plt.ylabel('Preference (%)', fontsize=14)
plt.title('The results of Rice/Wheat export', fontsize=20)

plt.legend()
plt.show()
```

## B. Seaborn

### 28. Menggunakan contoh dataset tips dari seaborn.

```python
import seaborn as sns

# gunakan dataset pemberian tips pada sebuah restoran
tips = sns.load_dataset("tips")
tips.info()
```

### 29. Menampilkan data total bill pada dataset tersebut menggunakan histogram

```python
# menampilkan data total bill yang di dapatkan
sns.histplot(data=tips, x="total_bill")
```

### 30. Ubahlah soal no. 29 menjadi Density Plot

```python
sns.kdeplot(data=tips, x="total_bill")
```

### 31. Menampilkan hubungan antara total bill dengan waktu makan

```python
# Menampilkan hubungan antara total bill dengan waktu makan
sns.kdeplot(data=tips, x="total_bill", hue="time", fill=True, alpha=0.6, linewidth=1.5)

plt.title("Hubungan Antara Total Bill dengan Waktu Makan")
plt.xlabel("Total Bill ($)")
plt.ylabel("Density")

plt.show()
```

### 32. Menampilkan hubungan pemberian tips terhadap total bill yang dibayarkan

```python
# Menampilkan hubungan pemberian tips terhadap total bill
sns.scatterplot(x="total_bill", y="tip", data=tips)

plt.title("Hubungan Antara Total Bill dan Tip")
plt.xlabel("Total Bill")
plt.ylabel("Tip")

plt.show()
```

### 33. Tampilkan hubungan besaran tips terhadap total bill yang dibayarkan sesuai jenis kelamin sehingga hasilnya sesuai dengan gambar di bawah ini

```python
# Menampilkan hubungan total bill, tip, jenis kelamin, dan ukuran grup
plt.figure(figsize=(8, 6))
sns.scatterplot(x="total_bill", y="tip", hue="sex", size="size", data=tips, sizes=(20, 200))

plt.title("Hubungan Antara Total Bill dan Tip")
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.legend(title="sex")

plt.show()
```

### 34. Menampilkan hubungan antar variable (numerik) pada dataset.

```python
sns.pairplot(data=tips)
plt.show()
```

### 35. Apa kecenderungan hubungan antara total bill dengan tips yang dibayarkan sesuai grafik tsb.? Jelaskan.

Berdasarkan pairplot yang ditampilkan, terdapat beberapa kecenderungan hubungan antara variabel total_bill, tip, dan size:

1. Hubungan antara total_bill dan tip

- Scatterplot antara total_bill dan tip menunjukkan pola sebaran yang cenderung naik. Ini mengindikasikan bahwa semakin besar total_bill, semakin besar juga tip yang diberikan.
- Hubungan ini tampak hampir linier positif, meskipun terdapat beberapa titik data di mana tip tidak terlalu meningkat secara proporsional terhadap total_bill.

2. Hubungan antara total_bill dan size (jumlah orang)

- Tidak terlihat pola yang sangat jelas, tetapi ada indikasi bahwa semakin besar jumlah orang (size), semakin besar pula total_bill.
- Namun, banyak titik data menunjukkan bahwa untuk jumlah orang yang sama, total_bill bisa bervariasi cukup luas.

3. Hubungan antara tip dan size

- Tidak ada hubungan yang terlalu kuat antara jumlah orang (size) dan jumlah tip.
- Dalam beberapa kasus, kelompok yang lebih besar mungkin memberikan tip yang lebih tinggi, tetapi ini tidak berlaku secara konsisten.
